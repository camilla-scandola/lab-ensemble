```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

Now perform the same as before:
- Feature Scaling
- Feature Selection


```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = spaceship.drop("Transported", axis=1)
y = spaceship["Transported"]

# split first
X_train, X_test, y_train, y_test = train_test_split(spaceship, y, test_size=0.2, random_state=42)

# identify numeric columns
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns

# scale numeric columns only
scaler = StandardScaler()

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test_scaled[numeric_cols] = scaler.transform(X_test[numeric_cols])
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
df_num = spaceship.select_dtypes(include=["float64", "int64"])

# Correlation matrix
corr = np.abs(df_num.corr())

# Mask upper triangle
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True

# Plot heatmap
plt.figure(figsize=(10, 10))
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(
    corr,
    mask=mask,
    vmax=1,
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.5},
    annot=corr.round(2)
)

plt.show()
```

```python
from sklearn.feature_selection import mutual_info_classif
import pandas as pd

# Select only numeric columns for X
X_num = X.select_dtypes(include=["float64", "int64"]).copy()

# Drop Nan values

X_num = X_num.fillna(X_num.median())

# Fill missing numeric values
X_num = X_num.fillna(X_num.median())

# Compute mutual information
mi = mutual_info_classif(X_num, y)

# Turn results into a Series for readability
mi = pd.Series(mi, index=X_num.columns)

# Sort descending (most important first)
mi.sort_values(ascending=False)
```

**Perform Train Test Split**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split your numeric features
X_train, X_test, y_train, y_test = train_test_split(X_num, y, test_size=0.2, random_state=42)

# SCALE (Standardization)
scaler = StandardScaler()
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.transform(X_test)
```

**Model Selection** - now you will try to apply different ensemble methods in order to get a better model

- Bagging and Pasting

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

bag_clf = BaggingClassifier(
    DecisionTreeClassifier(max_depth=20),
    n_estimators=100,
    max_samples=1000
)

bag_clf.fit(X_train_norm, y_train)

pred = bag_clf.predict(X_test_norm)

print("Accuracy:", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

- Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

clf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    random_state=42
)

clf.fit(X_train_norm, y_train)

pred = clf.predict(X_test_norm)

print("Accuracy:", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

- Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

# X_train_norm, X_test_norm, y_train, y_test

gb_clf = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,          
    random_state=42
)

# train model
gb_clf.fit(X_train_norm, y_train)

# evaluate model
pred = gb_clf.predict(X_test_norm)

print("Accuracy:", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

- Adaptive Boosting

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

base_tree = DecisionTreeClassifier(max_depth=3, random_state=42)

ada_clf = AdaBoostClassifier(
    estimator=base_tree,      
    n_estimators=100,
    learning_rate=0.1,
    random_state=42
)

ada_clf.fit(X_train_norm, y_train)

pred = ada_clf.predict(X_test_norm)

print("Accuracy:", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```
